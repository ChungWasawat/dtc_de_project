# transform   
### convert to datetime

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
df.dtypes
```

### convert gender number to string 

```python
def changeGender(num:int) -> str:
    if num == 0:
        return "unknown"
    elif num == 1:
        return "male"
    elif num == 2:
        return "female"
    
```

```python
df['gender'] = df['gender'].apply(changeGender)
```
